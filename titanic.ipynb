```python
import pandas as pd
import sklearn
import numpy as np
```

```python
df=pd.read_csv("./train.csv")
```

```python
df.head(5)
```

```python
X_train=df[["Pclass","SibSp","Parch"]]
Y_train=df["Survived"].values
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
X_train = X_train.apply(le.fit_transform)
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder,LabelBinarizer
y = LabelBinarizer().fit_transform(df.Sex)
```

```python
sex=pd.DataFrame(y,columns=["Sex"])
```

```python
X_train["Sex"]=sex
```

```python
X_train
```

```python
age=df['Age'].fillna(float(0)).values
```

```python
age=age.astype(int)
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(missing_values=0, strategy='mean')
```

```python
age=age.reshape(-1,1)
imp_mean.fit(age)
age=imp_mean.transform(age)
```

```python
age=age.astype(int)
age=pd.DataFrame(age,columns=["Age"])
X_train["Age"]=age
```

```python
X_train.head(15)
```

# Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rnd_clf = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1)
rnd_clf.fit(X_train[:750], Y_train[:750])
# print(X_train.shape,Y_train.shape)
```

```python
y_pred_rf = rnd_clf.predict(X_train[750:])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_train[750:],y_pred_rf)
```

# Using AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
ada_clf = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1), n_estimators=400,algorithm="SAMME.R", learning_rate=0.5)
ada_clf.fit(X_train[:750], Y_train[:750])
```

```python
y_pred_rf_2 = ada_clf.predict(X_train[750:])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_train[750:],y_pred_rf_2)
```

# Using Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gbclf = GradientBoostingClassifier(max_depth=2, n_estimators=5, learning_rate=1.0)
gbclf.fit(X_train[:750], Y_train[:750])
```

```python
y_pred_rf_3 = gbclf.predict(X_train[750:])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_train[750:],y_pred_rf_3)
```
